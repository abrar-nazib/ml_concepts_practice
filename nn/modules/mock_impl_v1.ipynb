# Basic Binary Classification Problem with Mock Dataset

```python
import torch
import torch.nn as nn
from torchinfo import summary
```

```python
class SimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1) # Input num_features, output 1
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, x):
        z = self.linear(x)
        y_pred = self.sigmoid(z)
        return y_pred
    
    # Internally __call__ function has been overridden so that calling model() calls forward() function with specified features
```

```python
# Create Fake dataset
features = torch.rand(10, 5)
print(features)
```

```python
# Create model
model = SimpleNN(features.shape[1])
```

```python
# Inference
# model.forward(features) will work but the recommended way is below
model(features)
```
