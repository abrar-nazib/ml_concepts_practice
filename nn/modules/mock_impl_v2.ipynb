# Model with hidden layers

```python
import torch
import torch.nn as nn
from torchinfo import summary
```

```python
class SimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.ReLU(),
            nn.Linear(3, 1),
            nn.Sigmoid()
        )
    
    
    def forward(self, x):
        out = self.network(x)
        return out
    
```

```python
# Create Fake dataset
features = torch.rand(10, 5)
print(features)
```

```python
# Create model
model = SimpleNN(features.shape[1])
```

```python
# Inference
# model.forward(features) will work but the recommended way is below
model(features)
```

```python
summary(model,input_size=(10, 5))
```
