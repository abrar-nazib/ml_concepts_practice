# Autograd Function

```python
import torch
import numpy as np
import pandas as pd
import seaborn as sns
import math
```

## Basic Examples

```python
# Example 1
x = torch.tensor(3.0, requires_grad=True) # Scalar with gradiation indication
y = x**2 # y will pick powerbackward method automatically
print(x, y)
y.backward() # Calculates the gradients
print(x.grad)

```

```python
# Example 2
x = torch.tensor(3.0, requires_grad=True) # Scalar with gradiation indication
y = x**2 # y will pick powerbackward method automatically
z = torch.sin(y)
print(x, y, z)
z.backward() # Calculates the gradients
print(x.grad)
```

## Implementation of perceptron using autograd

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

z = w*x + b
print(z)

# Calculate sigmoid
y_pred = torch.sigmoid(z)
print(y_pred)

# Calculate the loss
loss = torch.nn.functional.binary_cross_entropy(y_pred, y)

# Calculate the gradient
loss.backward()

print(w.grad, b.grad)
```

## Working with Vectors

```python
x = torch.tensor([1.0, 2.0, 3.0], requires_grad = True)
y = (x**2).mean()
print(y)
y.backward()
x.grad
```

## Gradient accumulation issue

- Per forward -> backward -> grad pass, gradient value accumulates

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2 # forward pass
y.backward() # backward pass
print(x.grad, x) # expected: 4
y = x ** 2
y.backward()
print(x.grad, x) # Gets added to gradient. Value = 8
x.grad.zero_() # Clear the gradient
y = x ** 2
y.backward()
print(x.grad, x) # Now it should be 4 again
```

## Disabling Gradient Tracking

```python
# Training mock behavior
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2
y.backward()
print(x.grad, x, y)

#  After training is done, reduce memory overhead of gradient tracking
x.requires_grad_(False)
y = x ** 2
print(x, y)

try:
    y.backward()
except Exception as e:
    print("Can't call backward() method now")
```

```python
# Second way of doing this
x1 = torch.tensor(2.0, requires_grad=True)
x2 = x1.detach()

y1 = x1 ** 2
y2 = x2 ** 2

print(y1, y2)
```

```python
# Best way of doing this
with torch.no_grad():
    y3 = x1 ** 2
    print(y3)
```
