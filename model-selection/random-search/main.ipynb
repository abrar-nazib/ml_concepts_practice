# RandomSearchCV

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Breast cancer dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
breast_cancer_dataset
```

```python
# Load data into dataframe
df = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
df['label'] = breast_cancer_dataset.target
df.head()
```

```python
# Check the shape
df.shape
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Feature:Label Split
X = df.drop(columns='label', axis='columns')
Y = df['label']
X.shape, Y.shape
```

```python
# Convert to numpy for easier processing
X = np.asarray(X)
Y = np.asarray(Y)
```

## Usage of RandomizedSearchCV

```python
# Create the pipeline
pipe = Pipeline([("scaler", StandardScaler()), ("svc", SVC())])

# Define the hyperparameters
param_grid = {
    'svc__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'svc__C': [0.1, 1, 10, 100] 
    # Small C = softer margin (more tolerance for misclassification)
    # large C = harder margin (penalizes misclassification more)
}
```

```python
# Grid Search
classifier = RandomizedSearchCV(pipe, param_grid, cv=5, n_jobs=-1)
    # n = -1 uses all CPU cores to accelerate the task
```

```python
# Run the training. A lot of iterations happen here. Each combination runs for 5 times
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# Best parameters
best_parameters = classifier.best_params_

# Highest accuracy
highest_accuracy = classifier.best_score_

print(f"Best Params: {best_parameters}, Highest Accuracy: {highest_accuracy}")  
```

```python
# Load results to pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```
